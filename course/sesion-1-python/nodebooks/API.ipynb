```python
class ExchangeAPI:
    def __init__(self, api_key):
        self.api_key=api_key
        self.url = 'http://api.exchangeratesapi.io/v1'
    def latest(self):
        url = f'{self.url}/latest?access_key={self.api_key}'
        resp = requests.get(url)
        if resp.ok:
            return resp.json()
        else:
            return None
    def convert(self, to='USD', amount=1):
        rates = self.latest()
        if rates:
            return rates['rates'][to] * amount
        else:
            return -1
```

```python
change = ExchangeAPI('d890db5fb56295adcfa689eff88e115b')
```

```python
change.latest()
```

```python
change.convert('GBP', 100)
```
